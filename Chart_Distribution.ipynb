```python
df = pd.read_csv("/content/gapminder(2007).csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df.head(20)
```

```python
#Country distribution
sns.countplot(x='country', data=df, palette='Set1')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x="continent",data=df, palette='rainbow')
plt.xticks(rotation=90)
plt.show()
```

#Findings
* Africa has sthe highest values because it has the highest bar above 50
* It is followed by Asia ---> Europe ---> Americas ---> Oceania in descending order


```python
#Barchart-countries vs life_exp
sns.barplot(x='continent',y='life_exp', data=df, palette='rainbow')
plt.xticks(rotation=90)
plt.show()
```

#Findings
* Oceania has the highest life expectency
* It is followed by Europe --> Asia --> Americas --> Africa
* Africa with the lowest life expectency

```python
df_countries=df.sort_values(by='population', ascending=False).head(20)
sns.barplot(x='country',y='population',data=df_countries, palette='rainbow')
plt.title('Top 20 Countries with Highest Population')
plt.xticks(rotation=90)
plt.show()
```

#Findings
* China and India have a combined population of over 2 billion
* Other countries have a population less than 0.4 billion

```python
df_countries=df.sort_values(by='population', ascending=False).head(20)
plt.bar(df_countries['country'],df_countries['population'])
plt.title('Top 20 Countries with Highest Population')
plt.xticks(rotation=90)
plt.show()
```

```python
#Stacked bar plot
#plt.bar(df_countries['country'],df_countries['population'], color='turquoise')
plt.bar(df_countries['country'], df_countries['life_exp'],bottom=df_countries[
  'population'], color='orange')
plt.title('Top 20 Countries with Highest Population')
plt.xticks(rotation=90)
plt.show()
```
